```python
hsv_img = cv2.cvtColor(original_img , cv2.COLOR_RGB2HSV)
h,s,v=cv2.split(hsv_img)
# s[h>165] = 135
# v[h>165] = 120
h[h<25] += 50
h[h>165] -=110
hsv_img[:,:,0] = h.astype(int)
hsv_img[:,:,1] = s
hsv_img[:,:,2] = v
img = cv2.cvtColor(hsv_img , cv2.COLOR_HSV2RGB)
plt.rcParams['figure.figsize'] = (20,10)
plt.subplot(1,2,1)
plt.imshow(original_img)
plt.subplot(1,2,2)
plt.imshow(img)
```
